L'ensemble des tâches planifiées enregistrer auprès d'une machine sont représentés par des clés de registres dont la valeur de chacune réside dans
HKLM\Software\Microsoft\Windows NT\CurrentVersion\Schedule\TaskCache.
Ainsi, on peut déterminer que les tâches sont créées en surveillant les ajouts de clé dans ce chemin.
Cela ne nous garantit pas qu'il s'agit de tâche créer à distance, mais cela nous permet d'identifier des tâches créées.
Il faut poursuivre ensuite avec la recherche sur les lignes de commandes.

```python
import pandas as pd
```

```python
data = pd.read_json("datasets/CONFIDENTIAL_remote_scheduled_task.json", lines=True)
```

```python
# https://www.socinvestigation.com/threat-hunting-using-windows-scheduled-task/
# https://redcanary.com/threat-detection-report/techniques/scheduled-task/
"""
Sysmon Event ID 1 - Relates to process creation and will log detailed information about the process that was started, including the command line, user context, and hashes of process-related files4
Event ID 4698 – A scheduled task was created: This event generates every time a new scheduled task is created.
Event ID 4699 – A scheduled task was deleted: This event generates every time a scheduled task was deleted.
Event ID 4700 – A scheduled task was enabled: This event generates every time a scheduled task is enabled.
Event ID 4701 – A scheduled task was disabled: This event generates every time a scheduled task is disabled.
Event ID 4702 – A scheduled task was updated: This event generates every time scheduled task was updated/changed.
Event ID 4688 – A new process has been created: This event generates every time a new process starts.
"""
```

```python
task_event_created_df = data.loc[
    (data['EventID'].isin([1, 4688,  4698, 4699,  4700, 4701, 4702])),
    ["EventID","Hostname", 'Message', 'ProcessId', 'CommandLine', 'ParentCommandLine']
]
# task_event_created_df.head
event_counts = task_event_created_df['EventID'].value_counts()
print(event_counts)

""" Result
1       15
4688    10
4698     1
Name: EventID, dtype: int64
"""

```

- Nous avons détecté 1 Event ID 4698 associé à la création d'une nouvelle tâche planifiée
- Nous avons détecté 15 sysmon event id et 10 event id 4688 associé à la creation de processus
- Nous allons maintenant explorer les lignes de commandes qui sont disponibles pour les events id 4688 et sysmon evnt id 1.

## Compréhension de LSASS
LSASS est un processus système Windows qui gère l'authentification des utilisateurs. Lorsqu'un utilisateur se connecte à un système Windows, LSASS vérifie les informations d'identification de l'utilisateur pour s'assurer qu'il est autorisé à accéder au système. Il est aussi responsable de la création, de la modification et de la suppression des comptes utilisateur. LSASS stocke les informations d'identification des utilisateurs, comme les mots de passe par exemple, et utilise des techniques de hachage pour les protéger.

```python
out_network.groupby('port').apply(lambda group: np.max((group['bytes_sent']-group['bytes_sent'].mean())/group['bytes_sent'].std())).sort_values(ascending=False).dropna().head(5)
```

##### Justification du choix du port
<font color='green'>
<h3>Réponse: </h3>
<p>
Le z-score est une mesure statistique qui représente le nombre de déviations standards par lesquelles la valeur d'une donnée individuelle diffère de la moyenne de l'ensemble des données. Un z-score élevé indique que la donnée est beaucoup plus éloignée de la moyenne des autres données, ce qui peut être un indicateur d'anomalie dans le contexte de la cybersécurité.

Pour calculer le z-score, on soustrait d'abord la moyenne de chaque donnée individuelle, puis on divise le résultat par l'écart-type de l'ensemble des données.

Dans notre contexte, un z-score de 20.647407 pour le port 124 suggère que les données émises sur ce port sont atypiques par rapport à l'ensemble des données émises sur tous les ports. Les données émises sur le port 124 dépassent de 21 déviations standards la moyenne. Cette valeur élevée du z-score pourrait indiquer que le port 124 est utilisé de manière "non conventionnelle", ce qui pourrait être associé à une activité malveillante, comme l'exfiltration de données.

On doit donc examiner de plus près les communications sur le port 124 pour comprendre pourquoi elles sont si éloignées de la norme. On devrait aussi vérifier si les communications sur ce port correspondent à des applications légitimes.
